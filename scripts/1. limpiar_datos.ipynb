## Converting Data Types

```python
df.info()
```

```python
category_columns = ["categoria",
                    "subcategoria",
                    "tg_region",
                    "equipo_asignado",
                    "grupo_asignado",
                    "tipo_de_problema",
                    "tipo_de_producto",
                    "severidad",
                    "grupo_que_resolvio",
                    "tipo"]
df["severidad"].unique().tolist()
```

```python
df["severidad"] = df["severidad"].astype(str)
df["severidad"] = df["severidad"].str.replace("1", "Critica").replace("2", "Alta").replace("3", "Media").replace("4", "Baja")
df["severidad"] = df["severidad"].str.replace("Critica - Crítica", "Critica").replace("4 - Baja", "Baja").replace("3 - Media", "Media")
df["severidad"].value_counts()
```

```python
for col in category_columns:
    df[col] = df[col].astype("category")

df.info()
```

```python
datetime_columns = ["inicio_falla", "cierre_falla", "fecha_asignacion", "fin_afectacion"]
for col in datetime_columns:
    df[col] = pd.to_datetime(df[col], format="%d/%m/%Y %H:%M:%S")
    
df["tiempo_de_cierre"] = pd.to_datetime(df["tiempo_de_cierre"], format="%d-%b-%y")

df.info()
```

```python
df.drop_duplicates(inplace=True)
```
